# Linear Discriminant Analysis

This notebook depicts how the Linear Discriminant Analysis Algorithm. This algorithm is supervised and will use some labelled data. One important thing about LDA is that the new features space always has set number of features which is equal to 1 less than the total distinct classes in the data

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from keras.datasets import mnist
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
```

## MNIST Dataset

For our purposes we will just use MNIST Dataset to get a sense of how LDA works, because it is a well labelled dataset. For the purposes of this section, we will use the full 28 X 28 dimension images as the input feature vectors

```
# Loading the data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# Reshaping the data so that each image is just one flat vector
X_train = X_train.reshape((-1, 28*28))
X_test = X_test.reshape((-1, 28*28))

m_train, n_train = X_train.shape
m_train, n_train
```

### 2 Class LDA

```
# Extracting the data which corresponds to the digits that we will be focusing on
digits = [0, 1]
X = X_train[(y_train == digits[0]) | (y_train == digits[-1])]
y = y_train[(y_train == digits[0]) | (y_train == digits[-1])]

# Finding the new number of samples
m, n = X.shape

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plotting the numbers based on their PCA Reduction into two dimensions
to_two_dim = PCA(2)
digit_features = to_two_dim.fit_transform(X)
sns.scatterplot(digit_features[:, 0], digit_features[:, 1], hue=y, ax=ax[0])

# Plotting the numbers based on their PCA Reduction into one dimension
to_one_dim = PCA(1)
digit_features = to_one_dim.fit_transform(X)
sns.scatterplot(digit_features[:, 0], [0]*len(digit_features[:, 0]), hue=y, ax=ax[1])
```
